Q9. Plot the polynomial function \( y = x^3 - 2x^2 + x - 1 \) for \( x \) values ranging from -3 to 3.

Q10. Create a line plot of the piecewise linear function \( f(x) = \begin{cases} -x & \text{if } x \leq 0 \\ x & \text{if } x > 0 \end{cases} \) for \( x \) values ranging from -2 to 2.

```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
#Q9: Plot the polynomial function y = x^3 - 2x^2 + x - 1 for x in the range of -3 to 3.
x = np.linspace(-3, 3, 1000)
y = [i**3 - 2*i**2 + i - 1 for i in x]

plt.plot(x, y)
plt.xlabel('Range of x from -3 to 3')
plt.ylabel('Values of Y')
plt.title('The Polynomial Function')
plt.grid(True)
plt.axvline(0)
plt.axhline(0)
plt.show()
```

```python
# Q10. Create a line plot of the piecewise linear function \( f(x) = \begin{cases} -x & \text{if } x \leq 0 \\ x & \text{if } x > 0 \end{cases} \) for \( x \) values ranging from -2 to 2.
x = np.linspace(-2, 2, 100)
y = [-i if i <= 0 else i for i in x]

plt.plot(x, y)
plt.xlabel('Range of x from -2 to 2')
plt.ylabel('Values of Y')
plt.title('f(x)')
plt.show()
```
